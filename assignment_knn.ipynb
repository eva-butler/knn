```python
import numpy as np 
import pandas as pd

heart_df = pd.read_csv('./data/heart_failure_clinical_records_dataset.csv')
print(heart_df.shape)
heart_df.describe()
```

no missing vals

```python
heart_df = heart_df.drop('time', axis =1)

heart_df.corr()
```

The vars with the highest coorelation to death event are serum_creatinine, ejection_fraction, and age. 

```python
vars = ['anaemia','diabetes','high_blood_pressure','sex','smoking']
for var in vars:
    print(heart_df.loc[:,[var,'DEATH_EVENT']].groupby(var).describe())
   
```

The only feature that has a higher proprtion of the population die when the variable takes on 1 rather than 0 is sex. However, they have similar means so probably wouldn't be that helpful. However, blood_pressure and anaemia do seem like they would be helpful predictors.

```python
y = heart_df['DEATH_EVENT']
vars = ['age','ejection_fraction','serum_creatinine','high_blood_pressure','anaemia', ]
X = heart_df.loc[:,vars]
```

```python
def maxmin(x):
    u = (x-min(x))/(max(x)-min(x))
    return u
X = X.apply(maxmin)
```

```python
np.random.seed(42) 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=100)
```

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
k_bar = 25
k_grid = np.arange(1,k_bar) # The range of k's to consider
SSE = np.zeros(k_bar) 

for k in range(k_bar):
    knn = KNeighborsRegressor(n_neighbors=k+1)
    predictor = knn.fit(X_train,y_train) 
    y_hat = knn.predict(X_test)
    SSE[k] = np.sum( (y_test-y_hat)**2 ) # Bug in sklearn requires .values

SSE_min = np.min(SSE) # highest recorded accuracy
min_index = np.where(SSE==SSE_min) 
k_star = k_grid[min_index] # Find the optimal value of k
print(k_star)

plt.plot(np.arange(0,k_bar),SSE) # Plot accuracy by k
plt.xlabel("k")
plt.title("optimal k:"+str(k_star)+', SSE:'+str(SSE_min))
plt.ylabel('SSE')
plt.show()
```

```python
X = heart_df.drop('DEATH_EVENT',axis=1)
np.random.seed(42) 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=100)


from sklearn.neighbors import KNeighborsRegressor

# Determine the optimal k:
k_bar = 100
k_grid = np.arange(1,k_bar) # The range of k's to consider
SSE = np.zeros(k_bar) 

for k in range(k_bar):
    knn = KNeighborsRegressor(n_neighbors=k+1)
    predictor = knn.fit(X_train,y_train) 
    y_hat = knn.predict(X_test)
    SSE[k] = np.sum( (y_test-y_hat)**2 ) # Bug in sklearn requires .values

SSE_min = np.min(SSE) # highest recorded accuracy
min_index = np.where(SSE==SSE_min) 
k_star = k_grid[min_index] # Find the optimal value of k
print(k_star)

plt.plot(np.arange(0,k_bar),SSE) # Plot accuracy by k
plt.xlabel("k")
plt.title("optimal k:"+str(k_star)+', SSE:'+str(SSE_min))
plt.ylabel('SSE')
plt.show()
```

In the end, the model with a fewer number of variables did a better job predicting. The one with more variables chose a k value of 60 while the model trained on fewer variables had a k value of 8. So the model with fewer variables actually did much better because the sse associated with the elbow of the model k=8 is lower.
